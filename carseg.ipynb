```python
##### preprocessingmax power columns
# handling nan values
default_value = '0 bhp @ 0 rpm'
dataset['Max Power'].fillna(default_value, inplace=True)

# spliiting bhp and rpm
dataset[['Horse_power','Engine_rpm']]=dataset["Max Power"].str.split('@',expand=True)

# extracting int value from columns
dataset['Horse_power'] = pd.to_numeric(dataset['Horse_power'].str.replace(' bhp', ''), errors='coerce')
dataset['Engine_rpm'] = pd.to_numeric(dataset['Engine_rpm'].str.replace(' rpm', ''), errors='coerce')

```

```python
# removing nan value from engine speed just remove it  
mean_engine_speed=dataset['Engine_rpm'].mean()
dataset['Engine_rpm'].fillna(mean_engine_speed,inplace=True)
```

```python
dataset.isnull().sum()
```

```python
default_value = '0 Nm @ 0 rpm'
dataset['Max Torque'].fillna(default_value, inplace=True)

# spliiting bhp and rpm
dataset[['Torque_power','Torque_rpm']]=dataset["Max Torque"].str.split('@',expand=True)

# extracting int value from columns
dataset['Torque_power'] = pd.to_numeric(dataset['Torque_power'].str.replace(' Nm', ''), errors='coerce')
dataset['Torque_rpm'] = pd.to_numeric(dataset['Torque_rpm'].str.replace(' rpm', ''), errors='coerce')

```

```python
# drop max power and max torque column
dataset.drop(columns="Max Power",inplace=True)
dataset.drop(columns="Max Torque",inplace=True)

dataset.isnull().sum()
```

```python
# removing cc from engine columns
dataset["Engine"] = dataset["Engine"].str.replace("cc", "").astype(float)

# 
```

```python
# handling nan value in engine
mean_engine_value = dataset["Engine"].mean()
dataset["Engine"].fillna(value=mean_engine_value, inplace=True)


```

```python
# handling  nan value in length,width,height
m_len=dataset["Length"].mean()
m_wid=dataset["Width"].mean()
m_height=dataset["Height"].mean()

dataset["Height"].fillna(m_height,inplace=True)

dataset["Width"].fillna(m_wid,inplace=True)

dataset["Length"].fillna(m_len,inplace=True)


```

```python
dataset.isnull().sum()
```

```python
# missing value in seating capacity
dataset["Seating Capacity"].fillna(dataset["Seating Capacity"].mean(),inplace=True)
```

```python
# missing value in fuel tank 
dataset["Fuel Tank Capacity"].fillna(dataset["Fuel Tank Capacity"].mean(),inplace=True)
```

```python
dataset["Seating Capacity"].fillna(dataset["Seating Capacity"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
np.random.seed(1)
# Count the occurrences of 'fwd', 'rwd', and 'awd' in the 'Drivetrain' column
drive_train_counts = dataset['Drivetrain'].value_counts()

# Generate random choices for all missing values using 'replace=True'
random_choices = np.random.choice(drive_train_counts.index, size=dataset['Drivetrain'].isnull().sum(), p=drive_train_counts.values / drive_train_counts.sum(), replace=True)

# Fill missing values with randomly chosen 'fwd', 'rwd', and 'awd'
dataset.loc[dataset['Drivetrain'].isnull(), 'Drivetrain'] = random_choices
```
